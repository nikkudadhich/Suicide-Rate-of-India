```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Suicides_in_India.csv")
```

```python
sns.set_palette(sns.color_palette(["#FF5733", "#FF5733", "#FF5733", "#FF5733", "#FF5733"]))
```

```python
# Visualization 1: Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Total', data=df)
plt.title('Total by State')
plt.xlabel('State')
plt.ylabel('Total')
plt.xticks(rotation=90)
plt.show()
```

```python
 #Visualization 2: Count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Year', data=df)
plt.title('Count by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 3: Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Type', y='Total', data=df)
plt.title('Total by Type')
plt.xlabel('Type')
plt.ylabel('Total')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 4: Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='Total', data=df)
plt.title('Total by Gender')
plt.xlabel('Gender')
plt.ylabel('Total')
plt.show()
```

```python
# Visualization 5: Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='Total', data=df)
plt.title('Total over Years')
plt.xlabel('Year')
plt.ylabel('Total')
plt.show()
```

```python
# Visualization 6: Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Total', data=df)
plt.title('Total over Years')
plt.xlabel('Year')
plt.ylabel('Total')
plt.show()
```

```python
 #Visualization 7: Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Total'], bins=20, kde=True, color='#FF5733')
plt.title('Total Histogram')
plt.xlabel('Total')
plt.ylabel('Frequency')
plt.show()
```

```python
# expenditure of 4 department

slices= [43.38,56.61]

# department name
depts=["women","male"]

#color for each department
cols=["cyan","gold"]

# create pie chart

plt.pie(slices, labels=depts, colors=cols,startangle=90, explode=(0,0.2), shadow=True, autopct="%.1f%%")

# set title and legend
plt.title("suicide case in india")
plt.legend()

# show the pie chart
plt.show()
```
